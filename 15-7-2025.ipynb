### Types of arguments:
1. Required positional arguments
2. default arguments
3. keywrod arguments
4. variable-length arguments

````python
def add(x,y):
  print(f"x:{x} y:{y}")
  return x+y
````

````python
add(y=20,x=30)
````

````python
def add(*arg):
  s = 0
  for i in arg:
    s += i
  return s
````

````python
print(add(10,20))
print(add(10,20,30))
print(add(10,20,30,40))
````

````python
def add(**l):
  return l['x']+l['y']

add(x=100,y=200,z=300,a=25)
````

````python
l = list(map(int,input().split()))
res = []
for i in l:
  res.append((i*10) + 2)
print(res)
````

````python
l = list(map(int,input().split()))
res = list(map(lambda x : (x*10) + 2,l))

print(res)
````

### List comprehension:
[value for_loop if_statement]

````python
l = list(map(int,input().split()))
res = [(i*10) + 2 for i in l]
print(res)
````

````python
l = list(map(int,input().split(",")))
res = []
for i in l:
  if i % 2 == 0:
    res.append(i)
print(res)
````

````python
l = list(map(int,input().split(",")))
res = [i for i in l if i % 2 == 0]
print(res)
````

````python
l = list(map(int,input().split(",")))
list(filter(lambda x : x %2 == 0,l))
````
