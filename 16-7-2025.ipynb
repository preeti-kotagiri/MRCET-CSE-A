```python
class A:
  x = 10
  def __init__(self,x,y):
    self.x = x
    self.y = y
  def __add__(self,var):
    return self.x + var.x,self.y+var.y
  def display(self):
    print("This is class A display method",self.x,self.y)
```

```python
a = A(10,20)
b = A(100,200)
```

```python
a + b
```

```python
from abc import ABC,abstractmethod
```

```python
class Person(ABC):
  @abstractmethod
  def role(self):
    pass
```

### stack implementation:

```python
class Stack:
  def __init__(self):
    self.s = []
  def is_empty(self):
    return len(self.s) == 0
  def size(self):
    if self.is_empty():
      return "stack is empty"
    return len(self.s)
  def push(self,val):
    self.s.append(val)
  def pop_item(self):
    if self.is_empty():
      return "stack is empty"
    return self.s.pop()
  def peek(self):
    if self.is_empty():
      return "stack is empty"
    return self.s[-1]
  def display(self):
    if self.is_empty():
      return "stack is empty"
    return self.s
```

```python
s = Stack()
```

```python
s.is_empty()
```

```python
s.size()
```

```python
s.push(10)
s.push(20)

```

```python
s.size()
```

```python
s.display()
```

```python
s.pop_item()
```

```python
s.display()
```
